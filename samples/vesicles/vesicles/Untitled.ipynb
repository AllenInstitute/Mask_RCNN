```python
import os
dirs = os.listdir('/allen/programs/braintv/workgroups/neuralcoding/Parastoo/threeD_test/')
dataset_path = '/allen/programs/braintv/workgroups/neuralcoding/Parastoo/threeD_test/'
images_source = '/allen/programs/braintv/workgroups/neuralcoding/Parastoo/EM_Data/test250/'
img_files = os.listdir(images_source)
```

```python
def make_img_path(images_source, img_id):
    '''Takes image source and id to make the correct path for the image'''
    
    if img_id in range(10):
        image_path = '{}250_00{}.png'.format(images_source, img_id)
    elif img_id in range(10, 100):
        image_path = '{}250_0{}.png'.format(images_source, img_id)
    else:
        image_path = '{}250_{}.png'.format(images_source, img_id)
        
    return image_path


for d in dirs[11:12]:
    img_id = int(d[-3:])
    pre_image = img_id - 1
    post_image = img_id + 1
    image_path = make_img_path(images_source, img_id)
    pre_image_path = make_img_path(images_source, pre_image)
    post_image_path = make_img_path(images_source, post_image)
    print(image_path)
    print(pre_image_path)
    print(post_image_path)
    image_dir = dataset_path + d 
    print(image_dir)
    pre_image_dir = image_dir + '/pre_image/'
    post_image_dir = image_dir + '/post_image/'
    os.mkdir(pre_image_dir)
    os.mkdir(post_image_dir)
    pre_img_target_path = make_img_path(pre_image_dir, pre_image)
    post_img_target_path = make_img_path(post_image_dir, post_image)
    print(pre_img_target_path)
    print(post_img_target_path)
    shutil.copyfile(pre_image_path, pre_image_path + '{}')
```
